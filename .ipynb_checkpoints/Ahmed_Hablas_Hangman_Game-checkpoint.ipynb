## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to win the game.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible. In this way,all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, we have given length of word + 2 chances. For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.

```python
# Libraries we need 
import random

# Words to play on 
someWords = '''apple banana mango strawberry orange grape pineapple apricot lemon coconut watermelon cherry papaya berry peach lychee muskmelon'''

someWords = someWords.split(' ')

# randomly choose a secret word from our "someWords" LIST.

# Printing game rules
print("A fruit is selected randomly by the program, you will guess the fruit character by charcter.\nYou win if you guess all the characters correctly!")

play_flag = True
while play_flag:
    word = random.choice(someWords)
#    word = 'berry'
    original = word
    chances = len(word) + 2
    final_string = ""
    indecies = [-1] * len(word)
    chars = ["*"] * len(word)
    first_str = ""
    second_str = ""
    flag = False
    flag_break = False
    print("You have", chances, "chances, use them wisely!")
    
    final_form_chars = ""
    
    for chance in range(chances):
        
        # Check if the player already wins!
#        not_final_string = final_string
#        final_form_chars = not_final_string.join(chars)
        
        if final_form_chars == original:
            print("You win! It is '", original + "'!")
            flag = True
            flag_break = True
            break
    
        # Take the next guess and play!
        while True:
            char = input("Enter your guessed character!\n")
            if len(char) == 1:
                break
            else:
                print("Please enter exactly 1 character!")
        if char in word:
            index = word.find(char)
            if index not in indecies:
                indecies[index] = index
                first_str = word[:index]
                second_str = word[(index+1):]
                chars[index] = char
                print(final_string.join(chars))
                word = first_str + "*" + second_str
                print("Good Guess!")
                
                not_final_string = final_string
                final_form_chars = not_final_string.join(chars)
                if final_form_chars == original:
                    flag = True
        else:
            print("Bad guess!\n")
            flag = False
    if flag == False:
        print("You lose! It is'", original + "'!")
    elif flag == True and flag_break == False:
        print("You win! It is '", original + "'!")
        
    play = input("Do you want to play again? Y for yes, any key for no\n")
    if play == 'y' or play == 'Y':
        play_flag = True
    else:
        play_flag = False
```
